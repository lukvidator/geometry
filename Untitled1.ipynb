```python
class Vector(Point):
    def __init__(self, *args):
        if len(args) == 1:
            self._coord = np.array(args[0], dtype=np.float64)
        elif len(args) == 2:
            types = list(map(type, args))
            if types == (Point, Point):
                self._coord = args[1]._coord - args[0]._coord
            elif types == (np.ndarray, np.ndarray):
                self._coord = args[1] - args[0]
            else:
                self._coord = np.array(args[1], dtype=np.float64) - np.array(args[0], dtype=np.float64)
        else:
            raise WrongTypeException(f"Can't init Vector with {args}")
    
    def __neg__(self):
        return Vector(-self._coord)
    
    def __add__(self, other):
        if type(other) == Point:
            return Point(self._coord + other._coord)
        elif type(other) == Vector:
            return Vector(self._coord + other._coord)
        else:
            raise WrongTypeException(f"Can't add an object of type {type(other).__name__} to {type(self).__name__}")
    
    def __sub__(self, other):
        if type(other) == Vector:
            return self.__add__(-other)
        else:
            raise WrongTypeException(f"Can't add an object of type {type(other).__name__} to {type(self).__name__}")
            
    def __mul__(self, other):
        return Vector(self._coord * other)
    
    def __truediv__(self, other):
        return Vector(self._coord / other)
    
    __rmul__ = __mul__
    __iadd__ = __add__
    __radd__ = __add__
    __isub__ = __sub__
    __imul__ = __mul__
    __itruediv__ = __truediv__
    
    def dot(self, other):
        return np.dot(self._coord, other)
    
    def is_orthog(self, other):
        return self.dot(other) == 0.0
    
    def is_collinear(self, other):
        return np.linalg.matrix_rank(np.array([self._coord, other._coord])) == 1
    
    def cross(self, other):
        return Vector(np.cross(self, other))
```

```python
def extract_coefs(equation, variables):
    assert type(variables) in (str, list, tuple)
    assert isinstance(equation, str)
    assert len(variables) == 3

    equation = equation.translate(str.maketrans({" ": ""}))  # deleting the whitespaces
    pattern = r"(?P<sign>[-+]?)(?P<coef>\d+(?:\.\d*)?|\.\d+)\*(?P<x>" + "|".join(variables) + ")"

    pattern_compiled = re.compile(pattern)
    matches = pattern_compiled.findall(equation)

    variables_found = {x[-1] for x in matches}

    if len(variables_found) != len(variables):  # if there are any missing variables
        variables_to_insert = set(variables).difference(variables_found)  # finding missing ones

        for x in variables_to_insert:
            matches.append(('', '0', x))

    def sort_criteria(item):
        nonlocal variables
        return variables.index(item[-1])

    matches.sort(key=sort_criteria)

    # for the last coefficient (free one)
    # removing all the terms except the free one
    deleting = [match[0] + match[1] + "*" + match[-1] for match in matches]
    equation = reduce(lambda res, to_delete: res.replace(to_delete, ""), deleting, equation)

    pattern = r"(?P<sign>[-+]?)(?P<coef>\d+(?:\.\d*)?|\.\d+)[^\*\.\d]"
    pattern_compiled = re.compile(pattern)

    match = pattern_compiled.findall(equation)
    if not match:
        match = [('', '0')]

    matches += match

    return np.array([float(x[0] + x[1]) for x in matches])
```

```python
def find_three_plane_points(coefs):
    index = coefs.nonzero()[0][0]
    others = np.concatenate([coefs[:index], coefs[index + 1:]])
    others = -others / coefs[index]
    instances = np.zeros([3, len(others) - 1])
    instances[0, 0] = 1.
    instances[1, 1] = 1.
    instances = np.hstack([instances, np.ones([3, 1])])
    c = np.array([[np.dot(others, inst)] for inst in instances])
    instances = instances[:, :-1]
    if index == 0:
        instances = np.hstack([c, instances])
    elif index == len(others) - 1:
        instances = np.hstack([instances, c])
    else:
        instances = np.split(instances, index, axis=1)
        instances.insert(index, c)
        instances = np.hstack(instances)
    return instances
```
